## Skewed data and Outlier Handling Approach:
We have skewed data in our data set.
we will first Log trasformation our data and check the outliers, if the data is skewed after the log trasformation we will handle the oultier by IQR method. 

Let me breif why i perfer for this approach:

### Log Transformation:

This helps to stabilize the variance and reduce the skewness of the data, making it more normally distributed. It can also reduce the impact of large outliers.
### IQR Method:

After log transformation, using the IQR method to detect and treat outliers helps to further ensure that the extreme values are handled appropriately. The IQR method is robust and less sensitive to the presence of outliers compared to methods based on mean and standard deviation.

### Features that need to treated with IQR method

```python
outlr_cols_log_af_tr = [  'Delay_from_due_date_log',
                        'Changed_Credit_Limit_log',
                        'Num_Credit_Inquiries_log',
                        'Outstanding_Debt_log',
                        'Credit_Utilization_Ratio_log',
                        'Total_EMI_per_month_log',
                        'Amount_invested_monthly_log',
                        'Monthly_Balance_log' ]
```

```python
{ outlr_cols_log_af_tr[0]: "lower_bound",
 outlr_cols_log_af_tr[1]: "lower_bound",
 outlr_cols_log_af_tr[2]: "lower_bound",
 outlr_cols_log_af_tr[3]: "lower_bound",
 outlr_cols_log_af_tr[4]: "upper_bound & upper_bound ",
 outlr_cols_log_af_tr[5]: "lower_bound & upper_bound",
 outlr_cols_log_af_tr[6]: "lower_bound & upper_bound",
  outlr_cols_log_af_tr[7]: "lower_bound & upper_bound"
}
```

```python
# using IQR method we will treat the outlier
# Upper threshold = Q3 + (1.5*IQR)
# Lower threshold = Q1 - (1.5*IQR) 
# IQR = Q3-Q1
# Q1 = df[col].quantile(0.25)
# Q3 = df[col].quantile(0.75)
```

```python
def iqr_lower_bound(df, col):
    Q1 = df[col].quantile(0.25)
    Q3= df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - (1.5*IQR)
    return lower_bound
    
def iqr_upper_bound(df, col):
    Q1 = df[col].quantile(0.25)
    Q3= df[col].quantile(0.75)
    IQR = Q3 - Q1
    upper_bound = Q3 + (1.5*IQR)
    return upper_bound
```

```python
df4 = df3.copy()
```

```python
df4.shape
```

```python
df4.drop(df4[ df4 [ outlr_cols_log_af_tr[0]] <= iqr_lower_bound( df4,outlr_cols_log_af_tr[0] ) ].index, inplace=True)

df4.drop(df4[ df4 [ outlr_cols_log_af_tr[1]] <= iqr_lower_bound( df4,outlr_cols_log_af_tr[1] ) ].index, inplace=True)

df4.drop(df4[ df4 [ outlr_cols_log_af_tr[2]] <= iqr_lower_bound( df4,outlr_cols_log_af_tr[2] ) ].index, inplace=True)

df4.drop(df4[ df4 [ outlr_cols_log_af_tr[3]] <= iqr_lower_bound( df4,outlr_cols_log_af_tr[3] ) ].index, inplace=True)

df4.drop( df4 [
         (  df4 [ outlr_cols_log_af_tr[4]] <= iqr_lower_bound( df4,outlr_cols_log_af_tr[4] ) ) |
         ( df4 [ outlr_cols_log_af_tr[4]] >= iqr_upper_bound( df4,outlr_cols_log_af_tr[4] ))
         
].index, inplace=True)

df4.drop( df4 [
         (  df4 [ outlr_cols_log_af_tr[5]] <= iqr_lower_bound( df4,outlr_cols_log_af_tr[5] ) ) |
         ( df4 [ outlr_cols_log_af_tr[5]] >= iqr_upper_bound( df4,outlr_cols_log_af_tr[5] ))
         
].index, inplace=True)

df4.drop( df4 [
         (  df4 [ outlr_cols_log_af_tr[6]] <= iqr_lower_bound( df4,outlr_cols_log_af_tr[6] ) ) |
         ( df4 [ outlr_cols_log_af_tr[6]] >= iqr_upper_bound( df4,outlr_cols_log_af_tr[6] ))
         
].index, inplace=True)

df4.drop( df4 [
         (  df4 [ outlr_cols_log_af_tr[7]] <= iqr_lower_bound( df4,outlr_cols_log_af_tr[7] ) ) |
         ( df4 [ outlr_cols_log_af_tr[7]] >= iqr_upper_bound( df4,outlr_cols_log_af_tr[7] ))
         
].index, inplace=True)

df4.reset_index(drop=True)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
def clustering(df,n):
    kmeans = KMeans(n_clusters=n,init='k-means++',n_init=10, random_state=42)
    clusters = kmeans.fit_predict(df)
    return clusters
```

```python
def clustering_metrics(df,c1,c2,c3,c4):
    silhouette_avg_c2 = silhouette_score(df, c1)
    silhouette_avg_c3 = silhouette_score(df, c2)
    silhouette_avg_c4 = silhouette_score(df, c3)
    silhouette_avg_c5 = silhouette_score(df, c4)   
    metrics = pd.DataFrame( [ {'SS_Cluster_2': silhouette_avg_c2, 
                               'SS_Cluster_3': silhouette_avg_c3,
                               'SS_Cluster_4': silhouette_avg_c4, 
                               'SS_Cluster_5': silhouette_avg_c5,                                                  
                              }])
    return metrics
```

```python
def cluster_ss_score(df,k):
    clusters_2 = clustering(df,k[0])
    clusters_3 = clustering(df,k[1])
    clusters_4 = clustering(df,k[2])
    clusters_5 = clustering(df,k[3])
    return clustering_metrics(df, clusters_2,clusters_3,clusters_4,clusters_5)
```
