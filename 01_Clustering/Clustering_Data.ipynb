```python
df1 = df.copy()
```

```python
# unique features are converted to string dtype to avoid in stats caluculation
df1['ID'] = df1['ID'].astype(str)
df1['Customer_ID'] = df1['Customer_ID'].astype(str)
df1['Age'] = df1['Age'].astype(str)
df1['SSN'] = df1['SSN'].astype(str)
```

```python
df1.info()
```

```python
df2 = df1.drop(['SSN','Month'],axis=1) # drop this unique features we need note have it in our dataset to train our model
df2
```

```python
# our dataset contains the details in month wise for the same person(the banking details duplicated), we will make it summarized based on the name.

def name_df(df, name):    
    df_name = df[df['Name'] == name]
    
    df_name1 = df_name.select_dtypes(include=[np.number]).agg('mean').to_frame(name='MeanValue').reset_index().T
    df_name1.columns = df_name1.iloc[0]
    df_name3 = df_name1.drop('index').reset_index(drop=True)
    
    df_name2 = df_name.select_dtypes(include='object').mode().iloc[0].reset_index().T
    df_name2.columns = df_name2.iloc[0]
    df_name4 = df_name2.drop('index').reset_index(drop=True)
    
    df_name5 = pd.concat([df_name3,df_name4],axis=1)
    return df_name5
```

```python
unique_names = df2['Name'].unique()
df_name_summary = pd.concat([name_df(df2,name) for name in unique_names], ignore_index=True)
df_name_summary # summarized dataset
```

```python
df_name_summary.to_csv("df_name_summary.csv",index=False)
```
